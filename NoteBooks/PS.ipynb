```python
plt.figure(figsize=[25, 10])

# Create the histogram
counts, bins, bars = plt.hist(x=data['event_location_district'], bins=range(len(data['event_location_district'].unique()) + 1), edgecolor='black', align='left')

# Annotate counts over the bars
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, count, int(count), ha='center', va='bottom', fontsize=12)

plt.xlabel('Event Location District')
plt.ylabel('Count')
plt.title('Histogram of Event Location District')
plt.xticks(rotation=45)
plt.show()
```
